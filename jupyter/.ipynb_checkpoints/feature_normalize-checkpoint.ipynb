## 比较torchvision， nn.functional的normalize的区别，数据是否处理的正确


```python
filename = '../datasets/L6N255_testing_400.csv'
L=6
sample = pd.read_csv(filename)

anderson =np.array(sample.iloc[:, 0:L+2])
anderson =  torch.tensor(anderson)
```

```python
torchvision_a = anderson[(None,) *2]
nor_a1 = transforms.Normalize(torchvision_a.mean(dim=2), torchvision_a.std(dim=2))(torchvision_a)
plt.plot(np.arange(8), nor_a1[0,0,0,:])
```

```python
nor_a2 = nn.functional.normalize(anderson, dim=0)       # wrong
nor_a3 = nn.functional.normalize(anderson, dim=1)
print(nor_a2.mean(), nor_a2.max(), nor_a2.std(dim=0), nor_a2.min())
print('next')
print(nor_a3.mean(), nor_a3.max(), nor_a3.std(dim=0), nor_a2.min())
```

```python
rows = 1
data = [ sample.iloc[:rows, :8], nor_a1[0,0, :rows,:], nor_a2[:rows,:],nor_a3[:rows,:]]
data = [ sample.iloc[:rows, :8], nor_a1[0,0, :rows,:], nor_a2[:rows,:],nor_a3[:rows,:]]

fig = plt.figure()
for j in range(rows):
    for i ,v in enumerate(data):
        axes = fig.add_subplot(j+1, len(data), j*len(data)+i+1)
        v = np.array(v)
        axes.plot(np.arange(8), v[j,:])
# fig.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)
    
```

# 从绘图中可以看到，在以列为绘画单位，使用torchvision, nn.functional.normalize(xx,dim=0),保持了原始的分布,这是正确的。

```python
cols = 1
plot_len = 20
data = [ sample.iloc[:, :cols], nor_a1[0,0, :,:cols], nor_a2[:,:cols],nor_a3[:,:cols]]

fig = plt.figure()
for j in range(cols):
    for i ,v in enumerate(data):
        axes = fig.add_subplot(j+1, len(data), j*len(data)+i+1)
        v = np.array(v)
        axes.plot(np.arange(plot_len), v[:plot_len,j])
# fig.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)
    
```
